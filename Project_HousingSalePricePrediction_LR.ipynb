```python
Q = pd.DataFrame(A.isna().sum(),columns=["CT"])
W = Q[Q.CT > 0].index
for i in W:
        if(A[i].dtypes == "object"):
            x = A[i].mode()[0]
            A[i] = A[i].fillna(x)
        else:
            x = A[i].mean()
            A[i] = A[i].fillna(x)
```

```python
X = A.drop(labels=['Id','SalePrice','LotArea','KitchenAbvGr','EnclosedPorch','MSSubClass','OverallCond','YrSold','LowQualFinSF','MiscVal','BsmtHalfBath','BsmtFinSF2','3SsnPorch','MoSold','PoolArea','ScreenPorch','BedroomAbvGr'],axis=1)
Y = A[['SalePrice']]
```

```python
Y.head(3)
```

```python
cat = []
con = []
for i in X.columns:
        if(X[i].dtypes == "object"):
            cat.append(i)
        else:
            con.append(i)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X1 = pd.DataFrame(ss.fit_transform(X[con]),columns=con)
X2 = pd.get_dummies(X[cat])
X = X1.join(X2)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=31)
```

```python
from statsmodels.api import add_constant,OLS
xconst = add_constant(xtrain)
```

```python
lm = OLS(ytrain,xconst)
model = lm.fit()
model.summary()
```

```python
model.pvalues
```

```python
W = pd.DataFrame(model.pvalues,columns=["pval"])
column_to_drop = W.sort_values(by="pval",ascending=False)[0:1].index[0]
print(column_to_drop)
X = X.drop(labels=[column_to_drop],axis=1)
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=31)
from statsmodels.api import add_constant,OLS
xconst = add_constant(xtrain)
lm = OLS(ytrain,xconst)
model = lm.fit()
model.summary()
```

```python
for i in range(0,110):
    W = pd.DataFrame(model.pvalues,columns=["pval"])
    column_to_drop = W.sort_values(by="pval",ascending=False)[0:1].index[0]
    print(column_to_drop)
    X = X.drop(labels=[column_to_drop],axis=1)
    from sklearn.model_selection import train_test_split
    xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=31)
    from statsmodels.api import add_constant,OLS
    xconst = add_constant(xtrain)
    lm = OLS(ytrain,xconst)
    model = lm.fit()
    model.summary()
```
